```python
#Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset

dataset = pd.read_csv("https://github.com/AidanFi/EnergyEfficiency/blob/main/ENB2012_data.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.drop(dataset.columns[dataset.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
dataset.head()
```

```python
dataset.columns=["Relative Compactness","Surface Area","Wall Area","Roof Area","Overall Height","Orientaion",
                   "Glazing Area","Glazing Area Dist","Heating Load","Cooling Load"]
dataset.head()
```

```python
dataset.isnull().any()
```

```python
dataset.dropna(subset = ["Relative Compactness"], inplace=True) 
```

```python
dataset.notna()
```

```python
dataset.dropna(axis=0, how="any", thresh=None, subset=None, inplace=False)
```

```python
dataset.isnull().any()
```

```python
dataset.describe()
```

```python
dataset.hist(figsize=(15,15))
plt.show()
```

```python
dataset.corr()
```

```python
#Setting the value for X and Y

x = dataset[['Relative Compactness', 'Surface Area', 'Wall Area', 'Roof Area', 'Overall Height', 'Orientaion',
                   'Glazing Area', 'Glazing Area Dist']]

y = dataset['Heating Load']
```

```python
#Splitting the dataset

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
#Fitting the Multiple Linear Regression model
# Our Target Variable is Heating Load

from sklearn.linear_model import LinearRegression

mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
#Intercept and Coefficient

print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```python
#Prediction of test set

y_pred_mlr= mlr.predict(x_test)

#Predicted values

print("Prediction for test set: {}".format(y_pred_mlr))
```

```python
#Actual value and the predicted value
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
mlr_diff.head()
```

```python
#Model Evaluation

from sklearn import metrics

meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))

print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
